```python
def makeCategoricalValues(df, columnName, skipColumns): 
    # skip converting if value type is integer
    if isinstance(df[columnName][0], np.int64) or columnName in skipColumns:
        print(columnName, 'ignored')
        return df[columnName]
    
    counts = df[columnName].value_counts()
    print ('counts', counts)
    to_replace = list(counts.keys())
    print ('to_replace', to_replace)
    replace_values = list(range(len(to_replace)))
    print ('replace_values', replace_values)
    return df[columnName].replace(to_replace, replace_values)
```
